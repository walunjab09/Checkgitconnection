```python
##tensorflow >2.0
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
```

```python
### Vocabulary size
voc_size=10000
```

```python
# One hot representation of sentences (each word is replaced by its position in selected vocabulary):
onehot_repr = [one_hot(word,voc_size) for word in sent]
onehot_repr
```

```python
# Word embedding
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
# Making same length for each sentence
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
# Selecting number of dimensions to be shown for each word
dim = 10
```

```python
# Creating ANN with an Embedded layer
model = Sequential()
model.add(Embedding(voc_size, dim, input_length = sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
# Converting each word to a vector of length 8 (user defined)
model.predict(embedded_docs)
```
